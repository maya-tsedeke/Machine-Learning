```python
#import packages
import numpy as np
#Let's start by making our feature set and the associated labels.
input_features = np.array([[0,0],[1,0],[0,1],[1,1]])
output_labels = np.array([[0,0,0,1]])
#four samples and One target output (expected result) (4,1)
labels = output_labels.reshape(4,1)
np.random.seed(42)
#The next step is to create our neural network's hyperparameters.
#two input and one out put (2,1)
weights = np.random.rand(2,1)
#using the random.seed function, the script will always produce the same random results.
bias = np.random.rand(1)
lr = 0.01
iterations = 5000
# Helper function: define activation function and its derivative

#activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))
    #calculates the derivative of the sigmoid function
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
#We are now prepared to train our neural network to be able to determine whether or not a person is healthy.
for epoch in range(iterations):
    inputs = input_features
      # feedforward step1
    X_W = np.dot(input_features, weights) + bias

    #feedforward step2
    z = sigmoid(X_W)


    # backpropagation step 1
    E = z - labels

    print(E.sum())

    # backpropagation step 2
    dcost_dprediction = E
    dprediction_dz = sigmoid_der(z)

    z_delta = dcost_dprediction * dprediction_dz

    inputs = input_features.T
    weights -= lr * np.dot(inputs, z_delta)

    for num in z_delta:
        bias -= lr * num
```

```python
# testing the Neural network
t1 = np.array([0,0])
t2 = np.array([1,0])
t3 = np.array([0,1])
t4 = np.array([1,1])

print('============================ Test result for Four samples ============================ \n')
print("Test Sample 1({}, {}) = {}".format(0, 0, sigmoid(np.dot(t1, weights) + bias)))
print("Test Sample 2({}, {}) = {}".format(1, 0, sigmoid(np.dot(t2, weights) + bias)))
print("Test Sample 3({}, {}) = {}".format(0, 1, sigmoid(np.dot(t3, weights) + bias)))
print("Test Sample 4({}, {}) = {}".format(1, 1, sigmoid(np.dot(t4, weights) + bias)))
```

```python
# importing Python library
import numpy as np

# Step Function
def unitStep(v):
	if v >0:
		return 1
	else:
		return 0

# Perceptron Model
def perceptronModel(x, w, b):
	v = np.dot(w, x) + b
	y = unitStep(v)
	return y
w = np.array([1, 1])
wNOR = np.array([-1, -1])
def NOR_Logic(x):
	bNOR=1
	return perceptronModel(x, wNOR, bNOR)
def AND_logic(x):
	bAND = -1.5
	return perceptronModel(x, w, bAND)
def OR_logic(x):
	bOR = -0.5
	return perceptronModel(x, w, bOR)
# testing the Perceptron Model
t1 = np.array([1, 1])
t2 = np.array([1, 0])
t3 = np.array([0, 1])
t4 = np.array([0, 0])


print('============================ AND Logic ============================ \n')
print("AND({}, {}) = {}".format(1, 1, AND_logic(t1)))
print("AND({}, {}) = {}".format(1, 0, AND_logic(t2)))
print("AND({}, {}) = {}".format(0, 1, AND_logic(t3)))
print("AND({}, {}) = {}".format(0, 0, AND_logic(t4)))
print('============================ OR Logic ============================ \n')
print("OR({}, {}) = {}".format(1, 1, OR_logic(t1)))
print("OR({}, {}) = {}".format(1, 0, OR_logic(t2)))
print("OR({}, {}) = {}".format(0, 1, OR_logic(t3)))
print("OR({}, {}) = {}".format(0, 0, OR_logic(t4)))
print('============================ NOR Logic ============================ \n')
print("NOR({}, {}) = {}".format(1, 1,NOR_Logic(t1)))
print("NOR({}, {}) = {}".format(1, 0,NOR_Logic(t2)))
print("NOR({}, {}) = {}".format(0, 1,NOR_Logic(t3)))
print("NOR({}, {}) = {}".format(0, 0,NOR_Logic(t4)))
```

```python
def NOTunitStep(v):
	if v >= 0:
		return 1
	else:
		return 0
# Perceptron Model
def NOTperceptronModel(x, w, b):
	v = np.dot(w, x) + b
	y = NOTunitStep(v)
	return y
def NOT_logic(x):
	return NOTperceptronModel(x, w=-1, b=0.5)
def XOR_network(x):
	gate_and = AND_logic(x)
	gate_not = NOT_logic(gate_and)
	gate_or = OR_logic(x)
	combined_x = np.array([gate_not, gate_or])
	output = AND_logic(combined_x)
	return output
print('============================ XOR Network Output ==================== \n')
print("XOR({}, {}) = {}".format(1, 1,XOR_network(t1)))
print("XOR({}, {}) = {}".format(1, 0,XOR_network(t2)))
print("XOR({}, {}) = {}".format(0, 1,XOR_network(t3)))
print("XOR({}, {}) = {}".format(0, 0,XOR_network(t4)))
```
